```python
#Number of courses per subject
df['subject'].value_counts()
```

```python
#using countplot to plot Number of courses per subject
sns.countplot(df['subject'])
plt.xlabel("Subjects")
plt.ylabel("No. of courses per Subject")
```

```python
#for which level udemy courses provide the courses?

df['level'].value_counts()
```

```python
sns.countplot(df['level'])
plt.xlabel("Level")
plt.ylabel("Udemy courses level")
```

```python
#Display the count of paid and free courses
df['is_paid'].value_counts()
```

```python
#which course has more lectures(free or paid)

df.groupby(['is_paid']).mean()
```

```python
df.groupby(['is_paid']).mean(numeric_only=True)  #This ensures that only numeric columns are used in the mean() calculation.
```

```python
#which courses has high no of subscribers (free or paid)
sns.barplot(x ='is_paid', y = 'num_subscribers', data = df)
```

```python
#which level has the highest number of subscribers

sns.barplot(x ='level', y = 'num_subscribers', data = df)
```

```python
#Find most popular course title
df[df['num_subscribers'].max()==df['num_subscribers']]['course_title']
```

```python
# display 10 most popular coursesas per no of subscribers

top_10 = df.sort_values(by ='num_subscribers', ascending=False).head(10)
```

```python
sns.barplot(x = 'num_subscribers',y = 'course_title', data = top_10)
```

```python
#Find the course having highest no of reviews

sns.barplot(x = 'subject', y='num_reviews' ,data = df)
```

```python
#does price affect no of reviews

df.columns
```

```python
plt.figure(figsize = (15,6))
sns.scatterplot(x = 'price', y = 'num_reviews', data = df)
```

```python
#find total no of courses related to python

len(df[df['course_title'].str.contains('python', case = False)])
```

```python
py = (df[df['course_title'].str.contains('python', case = False)]).sort_values(by = 'num_subscribers', ascending = False).head(10)
```

```python
#in which yr highest no of courses were posted

df['Year'] = df['published_timestamp'].dt.year
```

```python
sns.countplot(x='Year', data=df)
```
